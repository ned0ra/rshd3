```python
import pandas as pd
```

```python
rad = pd.read_csv('Road Accident Data.csv', delimiter=',')
wid = pd.read_csv('World Important Dates.csv', delimiter=',')
```

```python
rad.head()
```

```python
rad['Accident Date']=rad['Accident Date'].str.replace('/', '-')
```

```python
rad['Carriageway_Hazards'].unique()
```

```python
rad['Carriageway_Hazards'] = rad['Carriageway_Hazards'].fillna('No data')
rad.head()
```

```python
wid.head()
```

```python
wid['Date'] = pd.to_numeric(wid['Date'], errors='coerce').fillna(0).astype(int)
wid['Month']=wid['Month'].str.replace('Unknown','0')
```

```python
wid['Date'].unique()
```

```python
wid['Month'].unique()
```

```python
wid['Year'] = pd.to_numeric(wid['Year'].str.replace('\D',''), errors='coerce')
```

```python
wid['Year'].unique()
```

```python
rad.to_parquet('road_accident_data.parquet.gzip',compression='gzip')
wid.to_parquet('world_important_dates.parquet.gzip',compression='gzip')
```
